# カテゴリ値の前処理

```python
#C1=9、C4=7、C6=16はユニークな値がすくないのでカテゴリ変数にする
columns_to_convert = ['C1', 'C4', 'C6']

for col in columns_to_convert:
    df[col] = df[col].astype('category')


#ユニークな値がC2=171、C3=347、C5=223と多いので、出現頻度を計算し、それを数値として使用する
columns_to_encode = ['C2', 'C3', 'C5']

for col in columns_to_encode:
    freq = df[col].value_counts(normalize=True)  # 正規化された出現頻度
    df[f'{col}_freq'] = df[col].map(freq)

# クラスタリングアルゴリズムを使用して、自動的に出現頻度をグルーピングする

from sklearn.cluster import KMeans
import numpy as np

# KMeansで自動的にグループを生成
# 各列に対して適切なクラスタ数を設定
columns_to_group = {
    'C2_freq': 5,  # C2は5分割
    'C3_freq': 5,  # C3は5分割
    'C5_freq': 3   # C5は3分割
}

for col, n_clusters in columns_to_group.items():
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df[f'{col}_group'] = kmeans.fit_predict(df[[col]])
    df[f'{col}_group'] = df[f'{col}_group'].astype('category')  # カテゴリ型に変換


#df.drop(columns=['C2', 'C3', 'C5'], inplace=True)
```

# 数値変数の前処理
今回のデータは匿名化されたデータであり、元々の意味がカテゴリ情報である可能性が高いです。可視化してユニーク値の数が少ない、分布が均等である場合、カテゴリ変数に変換したのですが、うまくいきませんでした。分布から右テールの長い特徴量にのみ、対数変換を適用しました。

```python
import numpy as np

# 対数変換を適用したい列リスト
columns_to_log_transform = ['I5', 'I11', 'I12', 'I13', 'I14']

for col in columns_to_log_transform:
    # 欠損値を中央値で補完
    if df[col].isnull().any():
        print(f"{col} に欠損値が {df[col].isnull().sum()} 個含まれています。補完を行います。")
        df[col] = df[col].fillna(df[col].median())

    # 0以下の値を小さな正の値に置き換える（安全策）
    if (df[col] <= 0).any():
        print(f"{col} に 0 以下の値が含まれています。クリッピングを行います。")
        df[col] = df[col].clip(lower=1e-6)

    # 対数変換を適用
    df[col] = np.log1p(df[col])

    # 確認
    print(f"{col} の処理後: 欠損値数={df[col].isnull().sum()}, 最小値={df[col].min()}")
```

## OneHotEncodingを適用する

```python
from sklearn.preprocessing import OneHotEncoder
import joblib  # エンコーダーの保存用
import os
import pandas as pd

# カテゴリ列を再指定
categorical_columns = ['I1', 'C1', 'C4', 'C6', 'I3', 'I4',
                       'I6', 'I7', 'I8', 'I9', 'I10',
                       'C2_freq_group', 'C3_freq_group', 'C5_freq_group']

# 存在するカテゴリ列のみを対象
categorical_columns = list(set(categorical_columns).intersection(df.columns))

# カテゴリ型を明示する
for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype('category')

# One-Hot Encoding の初期化
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# カテゴリ列のエンコーディング
X_categorical = encoder.fit_transform(df[categorical_columns])

# エンコーダーを保存
joblib.dump(encoder, "one_hot_encoder.pkl")

# エンコーディングされたデータを DataFrame に変換
encoded_columns = encoder.get_feature_names_out(categorical_columns)
X_categorical_df = pd.DataFrame(X_categorical, columns=encoded_columns, index=df.index)

# 元データフレームからカテゴリ列を削除し、エンコーディング結果を結合
X = df.drop(columns=categorical_columns).reset_index(drop=True)
X = pd.concat([X, X_categorical_df.reset_index(drop=True)], axis=1)

# 保存先のディレクトリを指定
output_dir = "/content/drive/My Drive/signate/train"
os.makedirs(output_dir, exist_ok=True)  # ディレクトリが存在しない場合は作成

# 保存ファイル名を指定
output_file = os.path.join(output_dir, "train_1228.tsv")

# データセットを保存 (TSV形式)
X.to_csv(output_file, sep='\t', index=False)

print(f"データセットが保存されました: {output_file}")
```
